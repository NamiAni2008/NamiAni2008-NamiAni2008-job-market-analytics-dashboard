```python
import requests
from bs4 import BeautifulSoup
import mysql.connector
import time

# MySQL connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",             # change if needed
    password="root", # change to your MySQL password
    database="jobs_data"
)
cursor = conn.cursor()

headers = {"User-Agent": "Mozilla/5.0"}

# Define job roles
roles = ["data+analyst", "python+developer", "data+scientist"]
count_total = 0

for role in roles:
    print(f"\nğŸ” Scraping jobs for: {role.replace('+', ' ').title()}")

    for page in range(1, 4):  # scrape first 3 pages
        print(f"ğŸ“„ Page {page}")
        url = f"https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords={role}&txtLocation=&sequence={page}&startPage={page}"
        
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")

        jobs = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")
        print(f"â¡ï¸ Found {len(jobs)} jobs on page {page}")

        for job in jobs:
            title = job.header.h2.a.text.strip() if job.header and job.header.h2 else "No title"
            company_tag = job.find("h3", class_="joblist-comp-name")
            company = company_tag.text.strip() if company_tag else "No company"

            location_tag = job.find("ul", class_="top-jd-dtl clearfix")
            location = location_tag.li.text.strip() if location_tag and location_tag.li else "No location"

            experience_tag = job.find("li", class_="exper")
            experience = experience_tag.text.strip() if experience_tag else "Not mentioned"

            skills_tag = job.find("span", class_="srp-skills")
            skills = skills_tag.text.strip().replace("\r\n", "").replace("  ", "") if skills_tag else "Not listed"

            posted_tag = job.find("span", class_="sim-posted")
            posted_date = posted_tag.span.text.strip() if posted_tag and posted_tag.span else "No date"

            cursor.execute("""
                INSERT INTO jobs (title, company, location, experience, skills, posted_date)
                VALUES (%s, %s, %s, %s, %s, %s)
            """, (title, company, location, experience, skills, posted_date))

            count_total += 1
            print(f"âœ… Inserted: {title}")
            time.sleep(0.2)

        conn.commit()
        time.sleep(1)

print(f"\nğŸ‰ Successfully inserted {count_total} total job listings into MySQL!")

cursor.close()
conn.close()
```
